## Neural Networks

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neural_network import MLPClassifier
```

```python
iris = load_iris()

X = iris.data
y = iris.target

X.shape
```

```python
# Visualising the data
class_colors = np.array(["red","green","blue"])
plt.scatter(X[:,1],X[:,3],c=class_colors[y])
```

```python
model = MLPClassifier(max_iter=1000,hidden_layer_sizes=(10, 10, 10))
model.fit(X,y)
model.score(X,y)
```
