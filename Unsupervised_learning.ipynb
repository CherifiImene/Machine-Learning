## K-mean Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans 
```

```python
# Generate data
X, y = make_blobs(n_samples=100, centers = 3, cluster_std=0.5, random_state=0) #nb_features = 2 
plt.scatter(X[:,0], X[:, 1])
```

```python
# Training the model
model = KMeans(n_clusters=8)
model.fit(X)
```

```python
#Visualiser les Clusters
predictions = model.predict(X)
plt.scatter(X[:,0], X[:,1], c=predictions)
```

```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
y = iris.target

model = KMeans(n_clusters=3)
model.fit(X)

predictions = model.predict(X)
class_colors = np.array(["red","green","blue"])
plt.scatter(X[:,1], X[:,3], c=class_colors[predictions])
```
